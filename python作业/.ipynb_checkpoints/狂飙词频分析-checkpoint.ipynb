```python
#（错误版）
import jieba
import wordcloud
import matplotlib.pyplot as plt 
txt=open("/Users/zhengxiqiao/Documents/pythonlearning/part2.txt",encoding="gbk").read()
#print(txt[1:70000])
stopwords=[]
for line in open('/Users/zhengxiqiao/Documents/pythonlearning/情感分析/hit_stopwords.txt',encoding="UTF-8"):
    stopdict = line.rstrip()
    stopwords.append(stopdict) 
stopwords[1:10]
seg_txt = jieba.lcut(txt) 
result = [] 

for word in seg_txt:
    if word not in stopwords and len(word)>1:
        result.append(word)
        
result[1:30]
counts = {}     # 通过键值对的形式存储词语及其出现的次数

for word in result:
    if len(word) == 1:    # 单个词语不计算在内
        continue          #continue：结束当次循环，继续执行后续次数循环
    else:
        counts[word] = counts.get(word, 0) + 1    

items = list(counts.items())
items.sort(key=lambda x: x[1], reverse=True) 
print(items[0:30])
for i in range(30):
    word, count = items[i]
    print("{0:<5}{1:>5}".format(word, count))
```

```python
import jieba
import wordcloud
import matplotlib.pyplot as plt 
txt=open("/Users/zhengxiqiao/Documents/pythonlearning/part2.txt",encoding="gbk").read()
#print(txt[1:70000])
stopwords=[]
for line in open('/Users/zhengxiqiao/Documents/pythonlearning/情感分析/hit_stopwords.txt',encoding="UTF-8"):
    stopdict = line.rstrip()
    stopwords.append(stopdict) 
stopwords[1:10]
jieba.add_word("高启强")
jieba.add_word("高启盛")
seg_txt = jieba.lcut(txt) 
result = [] 
#print("自定义分词： " + "/ ".join(seg_txt)) 
for word in seg_txt:
    if word not in stopwords and len(word)>1:
        result.append(word)
result[1:30]
counts = {}    
for word in result:
    if len(word) == 1:   
        continue         
    else:
        counts[word] = counts.get(word, 0) + 1    

items = list(counts.items())
items.sort(key=lambda x: x[1], reverse=True) 
print(items[0:30])
for i in range(30):
    word, count = items[i]
    print("{0:<5}{1:>5}".format(word, count))
```
