```python
import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
#读入数据
txt=open("/Users/zhengxiqiao/Documents/乱七八糟文件/总结视频.txt",encoding="utf-8").read()
print(txt[1:2421])
words = jieba.lcut(txt)
counts = {} 
for word in words:#统计词频
    if len(word) == 1: 
         continue          
    else:
        counts[word] = counts.get(word, 0) + 1   
        items = list(counts.items())
items.sort(key=lambda x: x[1], reverse=True) 
print(items[0:19])
for i in range(19):
    word, count = items[i]
    print("{0:<5}{1:>5}".format(word, count))
txt_wordcloud = " ".join(words) #合并词列表
txt_wordcloud
w = WordCloud(
     font_path="/Library/Fonts/Arial Unicode.ttf",
    background_color="white",width=1000,height=880).generate(txt_wordcloud)

plt.imshow(w, interpolation="bilinear")
plt.axis("off")
plt.show()
```
