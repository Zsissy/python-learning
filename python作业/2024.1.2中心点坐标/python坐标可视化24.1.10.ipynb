```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
txt1= open('/Users/zhengxiqiao/Documents/pythonlearning/python作业/2024.1.2中心点坐标/flow02040000.txt').read()
#print(txt1)
listA=txt1.split()
list1=[]
for i in range(0,int(len(listA)/2)):
    list1.append(float(listA[2*i]))
    x=list1
list2=[]
for c in range(0,int(len(listA)/2)):
    list2.append(float(listA[2*c+1]))
    y=list2
fig = plt.figure(figsize = (20, 12))
fig, axes = plt.subplots(1,1)
plt.plot(x,y)
plt.show()
```
