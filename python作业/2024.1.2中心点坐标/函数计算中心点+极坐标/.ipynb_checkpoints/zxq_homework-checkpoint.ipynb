```python
def read_txt(path1):
    filetxt=open(path1).read()
    return filetxt
```

```python
def center_calculate(file1):
    listA=file1.split()
    listx=[]
    listy=[]
    for i in range(0,int(len(listA)/2)):
        listx.append(listA[2*i])
        listy.append(listA[2*i+1])
    sumx=0
    sumy=0
    for i in range(1,int(len(listx))+1):
        x1=float(listx[i-1])
        sumx=sumx+x1
        y1=float(listy[i-1])
        sumy=sumy+y1
    cx=sumx/len(listx)
    cy=sumy/len(listy)
    return cx,cy,listx,listy
```

# 细节整理：
1.append的用法
正确用法是：
~~~
listr.append((r,theta))
~~~
错误用法是：
~~~
listr=listr.append((r,theta))
~~~
区别在于，正确用法无listr=，而是直接append

2.listx[i]输出的不一定是数值，利用它进行运算的时候要确定它的数据形式
如果是字符串，转化成整数：
~~~
int(listx[i])
~~~
如果是字符串，转化成小数：
~~~
float(listx[i])
~~~
