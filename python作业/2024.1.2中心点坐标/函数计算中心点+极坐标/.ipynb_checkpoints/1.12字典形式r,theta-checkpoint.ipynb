```python
def read_txt(path1):
    filetxt=open(path1).read()
    return filetxt
```

```python
def center_calculate(file1):
    listA=file1.split()
    listx=[]
    listy=[]
    for i in range(0,int(len(listA)/2)):
        listx.append(listA[2*i])
        listy.append(listA[2*i+1])
    sumx=0
    sumy=0
    for i in range(1,int(len(listx))+1):
        x1=float(listx[i-1])
        sumx=sumx+x1
        y1=float(listy[i-1])
        sumy=sumy+y1
    cx=sumx/len(listx)
    cy=sumy/len(listy)
    return cx,cy,listx,listy
```
